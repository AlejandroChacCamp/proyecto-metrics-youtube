## **Ejercicio 2: Lista de Pokémon (paginación)**
La PokéAPI también tiene un endpoint para listar Pokémon: https://pokeapi.co/api/v2/pokemon?limit=10&offset=0.

    - limit: número de resultados por página.

    - offset: desplazamiento (cuántos resultados saltar).

Prueba a obtener los primeros 10 Pokémon. La respuesta no trae todos los detalles, solo nombres y una URL para obtener los detalles de cada uno. Esto es común en APIs: primero obtienes una lista resumida y luego, si quieres más detalles, haces una petición adicional a la URL proporcionada.

Tarea: Obtén los primeros 10 Pokémon y, para cada uno, imprime su nombre. Luego, elige uno (por ejemplo, el primero) y haz una segunda petición a su URL (url que viene en el campo url) para obtener sus detalles completos.

**Reflexión:**
¿Qué ventaja tiene este enfoque de dar primero una lista resumida y luego los detalles?

¿Cómo se relaciona esto con lo que podríamos encontrar en la API de YouTube? (pista: los canales tienen listas de videos, y cada video tiene detalles)

```python
import json
import requests
```

```python
lista_pokemon = requests.get("https://pokeapi.co/api/v2/pokemon?limit=10&offset=0")

datos_lista = lista_pokemon.json()

#print(datos_lista)

datos_lista_formateados = json.dumps(datos_lista, indent=2)
print(datos_lista_formateados)
```

```python
for clave, valor in datos_lista.items():
    print("key: ", clave)
    #print("valor: ", valor)
    print("Type: ", type(valor))
```

```python
count = 0
for i in datos_lista["results"]:
    nombre = i["name"]
    count += 1
    print(f"Nombre del pokemon: {nombre}, número de orden: {count}")
```

```python
pokemon_particular = datos_lista["results"][0]["url"]
print(pokemon_particular)

datos_bulbasaur_raw = requests.get(pokemon_particular)

datos_bulbasaur = datos_bulbasaur_raw.json()

"""
datos_bulbasaur_formateados = json.dumps(datos_bulbasaur, indent=2)
print(datos_bulbasaur_formateados)
"""
nombre = datos_bulbasaur["name"]
print("nombre:", nombre)
altura = datos_bulbasaur["height"]
print("altura:", altura)
peso = datos_bulbasaur["weight"]
print("peso:", peso)


lista_movimientos = []
for i in datos_bulbasaur["moves"]:
    movimiento = i["move"]["name"]
    lista_movimientos.append(movimiento)

print(f"Lista de movimeintos: {lista_movimientos}")
```
