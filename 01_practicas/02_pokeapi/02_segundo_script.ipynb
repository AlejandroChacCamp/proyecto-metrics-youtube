```python
lista_paginada = []

url_actual = "https://pokeapi.co/api/v2/pokemon?limit=10&offset=0"

for i in range(3):     
    
    next_page_json = requests.get(url_actual)
    resultados_next_page = next_page_json.json()

    resultados = resultados_next_page["results"]

    for elemento in resultados:
        nombre = elemento["name"]
        #print(nombre)
        lista_paginada.append(nombre)

    url_actual = resultados_next_page["next"]
    print(url_actual)
    print(f"final iteración {i}")
print(lista_paginada)
    


#pokemones_paginados = json.dumps(pokemones_paginados, indent =2)
#print(pokemones_paginados)
```

```python
lista_paginada = []

url_actual = "https://pokeapi.co/api/v2/pokemon?limit=100&offset=0"

while url_actual is not None:     
    
    next_page_json = requests.get(url_actual)
    resultados_next_page = next_page_json.json()

    resultados = resultados_next_page["results"]

    for elemento in resultados:
        nombre = elemento["name"]
        #print(nombre)
        lista_paginada.append(nombre)

    url_actual = resultados_next_page["next"]
    print(url_actual)
print(lista_paginada)
    


#pokemones_paginados = json.dumps(pokemones_paginados, indent =2)
#print(pokemones_paginados)
```
